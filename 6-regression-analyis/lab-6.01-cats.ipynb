## Exercise 1: Relation between heart weight and body weight in Cats

### Part 1

Perform a linear regression analysis on the variables body weight (`Bwt`, dependent variable) and heart weight (`Hwt`, independent variabele).


1. Import the dataset `Cats.csv`.


```python
cats = pd.read_csv('https://raw.githubusercontent.com/HoGentTIN/dsai-labs/main/data/Cats.csv')
# Show the first few records of the Data Frame
cats.head()
```

```python
katers = cats[cats.Sex == 'M']
```

```python
katinnen = cats[cats.Sex == 'F']
```

2. Draw a scatter plot of both variables.


```python
sns.regplot(data=cats, x=cats.Hwt, y=cats.Bwt);
```

3. Calculate and draw the regression line.


```python
b1, b0 = np.polyfit(x=cats.Hwt, y=cats.Bwt, deg=1)
print(f'y = {b0:.2f} + {b1:.2f}x')
```

```python
b1, b0 = np.polyfit(x=katers.Hwt, y=katers.Bwt, deg=1)
print(f'y = {b0:.2f} + {b1:.2f}x')
```

```python
b1, b0 = np.polyfit(x=katinnen.Hwt, y=katinnen.Bwt, deg=1)
print(f'y = {b0:.2f} + {b1:.2f}x')
```

4. Calculate the correlation coefficient and the coefficient of determination.

```python
cor = np.corrcoef(x=cats.Hwt, y=cats.Bwt)[0, 1]
print(f'Correlation coefficient R: {cor}, R^2: {cor**2}')
```

```python
cor = np.corrcoef(x=katers.Hwt, y=katers.Bwt)[0, 1]
print(f'Correlation coefficient R: {cor}, R^2: {cor**2}')
```

```python
cor = np.corrcoef(x=katinnen.Hwt, y=katinnen.Bwt)[0, 1]
print(f'Correlation coefficient R: {cor}, R^2: {cor**2}')
```

5. Interpret the results from the previous steps.

```python
#Cats:
R > 0 => het is een stijgend verband
R > 0.5 => het is een sterk verband
R^2 ~ 0.64 => 64% van de variatie van de afh. variable wordt verklaard door de onafh veranderlijke
```

```python
#Katers (Male cats)
R > 0 => het is een stijgend verband
R > 0.5 => het is een sterk verband
R^2 ~ 0.64 => 64% van de variatie van de afh. variable wordt verklaard door de onafh veranderlijke
```

```python
#Katinnen (Female cats)
R > 0 => het is een stijgend verband
R ~ 0.53 => het is een gemiddeld verband
R^2 ~ 0.28 => 28% van de variatie van de afh. variable wordt verklaard door de onafh veranderlijke
```
